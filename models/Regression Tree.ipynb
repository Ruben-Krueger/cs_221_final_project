```python
import os
import numpy as np
import sklearn
import pandas as pd
import collections

# Model taken from https://towardsdatascience.com/https-medium-com-lorrli-classification-and-regression-analysis-with-decision-trees-c43cdbc58054
```

```python
## Import and prepare data

df=pd.read_csv('data_full.csv')

#print(df) #Uncomment to see imported data
```

```python
## Features taken from Ruben's NN model + baseline

proteins = df['protein']

X = df.drop(['tm'], axis=1)
Y = df['tm']


for index, row in X.iterrows():
    sequence = row[1]
    counts = collections.Counter(sequence)
    
    # add (E + K)/(Q + H) ratio feature
    if (counts['Q'] + counts['H']) != 0:
        df.at[index, '(E + K)/(Q + H)'] = (counts['E'] + counts['K']) /(counts['Q'] + counts['H'])
    else:
        df.at[index, '(E + K)/(Q + H)'] = 0   
    
    # Length feature
    df.at[index, 'Length'] = len(sequence)
    
# Individual counts for each aminoacid features
for i in range(len(df.index)):
    counter = collections.Counter(df.iloc[i, 2])
    for key in counter.keys():
        df.at[i, key] = counter[key]
df = df.fillna(0)


print(df)

```
